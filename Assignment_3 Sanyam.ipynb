```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
T = 1.0       # Total time
N = 100       # Number of steps
dt = T / N    # Time step size
t = np.linspace(0, T, N + 1)  # Time grid

# Generate Brownian increments
dW = np.random.normal(loc=0.0, scale=np.sqrt(dt), size=N)

# Build Wiener process
W = np.zeros(N + 1)
W[1:] = np.cumsum(dW)

# Plot result
plt.figure(figsize=(10, 4))
plt.plot(t, W, label='Wiener Process Path')
plt.title("Simulated Path of Standard Brownian Motion")
plt.xlabel("Time t")
plt.ylabel("W(t)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
T = 1.0        # Total time
N = 100        # Number of time steps
dt = T / N     # Time step size
t = np.linspace(0, T, N + 1)   # Time vector

alpha = 0.1    # Drift term
sigma = 0.2    # Volatility
S0 = 100       # Initial value of the process

plt.figure(figsize=(10, 5))

# Simulate 5 paths
for i in range(5):
    # Simulate Brownian motion path
    dW = np.random.normal(loc=0.0, scale=np.sqrt(dt), size=N)
    W = np.zeros(N + 1)
    W[1:] = np.cumsum(dW)

    # Compute GBM path
    S = S0 * np.exp(sigma * W + (alpha - 0.5 * sigma**2) * t)

    # Plot this path
    plt.plot(t, S, label=f"Path {i+1}")

# Final plot formatting
plt.title("5 Paths of Geometric Brownian Motion")
plt.xlabel("Time t")
plt.ylabel("S(t)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np

# Parameters
s = 0.4
t = 0.7
n_paths = 10000

# Simulate W(s) and W(t)
Ws = np.random.normal(0, np.sqrt(s), size=n_paths)
Wt = Ws + np.random.normal(0, np.sqrt(t - s), size=n_paths)

# Compute expectation
expected_value = np.mean(Ws * Wt)

print(f"Empirical E[W(s)*W(t)]  {expected_value:.4f}")
print(f"Theoretical min(s,t) = {min(s,t)}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
s = 0.4
t = 0.7
n_samples = 10000

# Simulate W(s) and W(t)
Ws = np.random.normal(0, np.sqrt(s), size=n_samples)
Wt = Ws + np.random.normal(0, np.sqrt(t - s), size=n_samples)

# Compute the increment
dW = Wt - Ws

# Plot histogram and compare to theoretical normal
plt.hist(dW, bins=100, density=True, alpha=0.6, label='Simulated')
x = np.linspace(-1, 1, 200)
pdf = (1 / np.sqrt(2 * np.pi * (t - s))) * np.exp(-x**2 / (2 * (t - s)))
plt.plot(x, pdf, 'r', label='Theoretical N(0, t-s)')
plt.title("Distribution of W(t) - W(s)")
plt.xlabel("Increment")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np

# Simulate many paths of Brownian motion
N = 10000
s, t = 0.4, 0.7
Ws = np.random.normal(0, np.sqrt(s), size=N)
Wt = Ws + np.random.normal(0, np.sqrt(t - s), size=N)

# Empirical estimate of E[Wt | Ws]
print(f"Empirical E[Wt | Fs]  {np.mean(Wt - Ws):.4f}")
print(f"Theoretical: E[Wt - Ws] = 0, so E[Wt | Fs] = Ws")
```
